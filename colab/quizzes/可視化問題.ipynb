## 1問目
使うcsvはhist_data.csvです。下にコードを書いてください。

問1.png

```python
data1 = pd.read_csv("matplotlib_quiz/hist_data.csv")
data1["aデータ"].hist(bins = 30, label = "aデータ", color='orange', alpha = 0.5)
data1["bデータ"].hist(bins = 30, label = "bデータ", color='turquoise', alpha = 0.5)
plt.title("aデータとbデータの温度に関する度数分布", fontsize = 10)
plt.xlabel("温度℃", fontsize = 10)
plt.ylabel("度数", fontsize = 10)
plt.grid(False)
plt.legend()
```

## 2問目
使うcsvはscatter_data.csvです。下にコードを書いてください。

問2.png

```python
import seaborn as sns

data2 = pd.read_csv("matplotlib_quiz/scatter_data.csv")
```

```python
data2
```

```python
class0 = data2.loc[data2["class"] == 0]
class1 = data2.loc[data2["class"] == 1]
class2 = data2.loc[data2["class"] == 2]
plt.title("クラス0～2のlengthとwidthに関する散布図")
plt.xlabel("length (cm)")
plt.ylabel("width (cm)")
plt.grid(True)
plt.legend(loc = "upper right")
x = "length (cm)"
y = "width (cm)"
plt.scatter(x = class0["length (cm)"], y = class0["width (cm)"], s = 50, c = "r",
                      marker = "o", label="class0")
plt.scatter(x = class1["length (cm)"], y = class1["width (cm)"], s = 50, c = "b",
                      marker = "o", label="class1")
plt.scatter(x = class2["length (cm)"], y = class2["width (cm)"], s = 50, c = "g",
                      marker = "o", label="class2")
plt.legend(loc = "upper right")
```

```python
data3 = pd.read_csv("matplotlib_quiz/line_data.csv")
data3
```

```python
data3.plot(x = "期間", y = "件数", kind = "line", title = "期間ごとの件数の推移",
           marker = "o", rot = 45, xlabel = "期間", ylabel = "件数", legend = False, grid = True)
```

```python
data3.plot(x = "期間", y = "費用", kind = "bar", title = "期間ごとの費用の推移",
           rot = 45, xlabel = "期間", ylabel = "費用(円)", width = 0.8, legend = False, grid = True)
```
